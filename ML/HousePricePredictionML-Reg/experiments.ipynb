##Missing Values 

```python
na_col = [col for col in df.columns if df[col].isna().sum() > 0]
for col in na_col:
    print(f"{col}: {df[col].isna().sum()} missing values")
```

```python
for col in na_col:
    if df[col].dtype in ['int64', 'float64']:
        df[col].fillna(df[col].mean(), inplace=True)  # Numerical columns
    else:
        df[col].fillna('NotAvail', inplace=True)  # Categorical columns

    
```

```python
print("\nMissing values after filling:")
print(df.isna().sum().sum())  # Should be 0
```

```python
cat_cols = [col for col in df.columns if df[col].dtype == 'object']
```

```python
cardinality = {col: df[col].nunique() for col in cat_cols}
low_cardinality_cols = [col for col in cat_cols if cardinality[col] <= 10]
high_cardinality_cols = [col for col in cat_cols if cardinality[col] > 10]
```

```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(df, test_size=0.2, random_state=42)
```

```python
df_train
```

# for handling low cardinal features with One hot encoding

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler

encoder = OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore')
df_train_ohe = encoder.fit_transform(df_train[low_cardinality_cols])
df_test_ohe = encoder.transform(df_test[low_cardinality_cols])

df_train_ohe = pd.DataFrame(df_train_ohe, columns=encoder.get_feature_names_out(low_cardinality_cols))
df_test_ohe = pd.DataFrame(df_test_ohe, columns=encoder.get_feature_names_out(low_cardinality_cols))

# Concatenate OHE results and drop original categorical columns
df_train = pd.concat([df_train.reset_index(drop=True), df_train_ohe], axis=1).drop(columns=low_cardinality_cols)
df_test = pd.concat([df_test.reset_index(drop=True), df_test_ohe], axis=1).drop(columns=low_cardinality_cols)
```

##Target Encoding


```python
def target_encode_with_smoothing(df_train, df_test, categorical_cols, target_col, m=5):
    global_mean = df_train[target_col].mean()
    smoothed_means = {}
    
    for col in categorical_cols:
        means = df_train.groupby(col)[target_col].mean()
        counts = df_train.groupby(col)[target_col].count()
        
        smoothed_col_means = (counts * means + m * global_mean) / (counts + m)
        smoothed_means[col] = smoothed_col_means
        
        df_train[f'{col}_encoded'] = df_train[col].map(smoothed_col_means).fillna(global_mean)
        df_test[f'{col}_encoded'] = df_test[col].map(smoothed_col_means).fillna(global_mean)
    
    df_train.drop(columns=categorical_cols, inplace=True)
    df_test.drop(columns=categorical_cols, inplace=True)
    
    return df_train, df_test

df_train, df_test = target_encode_with_smoothing(df_train, df_test, high_cardinality_cols, 'SalePrice', m=5)
```

## Standard Scaler

```python
num_cols = [col for col in df.columns if df[col].dtype in ['int64', 'float64'] and col != 'SalePrice']
scaler = StandardScaler()
df_train[num_cols] = scaler.fit_transform(df_train[num_cols])
df_test[num_cols] = scaler.transform(df_test[num_cols])
```

```python
X_train = df_train.drop(columns=['SalePrice','Id'])
y_train = df_train['SalePrice']
X_test = df_test.drop(columns=['SalePrice','Id'])
y_test = df_test['SalePrice']
```

## Tranining


```python
# Basic Import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
# Modelling
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression, Ridge,Lasso
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.model_selection import RandomizedSearchCV
```

```python
def evaluate_model(true, predicted):
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    rmse = np.sqrt(mean_squared_error(true, predicted))
    r2_square = r2_score(true, predicted)
    return mae, rmse, r2_square
```

```python
models = {
    "Linear Regression": LinearRegression(),
    "Lasso": Lasso(),
    "Ridge": Ridge(),
    "K-Neighbors Regressor": KNeighborsRegressor(),
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest Regressor": RandomForestRegressor(),
    "AdaBoost Regressor": AdaBoostRegressor()
}
```

```python
model_list = []
r2_list =[]

for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train, y_train) # Train model

    # Make predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    # Evaluate Train and Test dataset
    model_train_mae , model_train_rmse, model_train_r2 = evaluate_model(y_train, y_train_pred)

    model_test_mae , model_test_rmse, model_test_r2 = evaluate_model(y_test, y_test_pred)

    
    print(list(models.keys())[i])
    model_list.append(list(models.keys())[i])
    
    print('Model performance for Training set')
    print("- Root Mean Squared Error: {:.4f}".format(model_train_rmse))
    print("- Mean Absolute Error: {:.4f}".format(model_train_mae))
    print("- R2 Score: {:.4f}".format(model_train_r2))

    print('----------------------------------')
    
    print('Model performance for Test set')
    print("- Root Mean Squared Error: {:.4f}".format(model_test_rmse))
    print("- Mean Absolute Error: {:.4f}".format(model_test_mae))
    print("- R2 Score: {:.4f}".format(model_test_r2))
    r2_list.append(model_test_r2)
    
    print('='*35)
    print('\n')
```

```python
pd.DataFrame(list(zip(model_list, r2_list)), columns=['Model Name', 'R2_Score']).sort_values(by=["R2_Score"],ascending=False)
```

```python
from sklearn.model_selection import GridSearchCV

# Define model
rf = RandomForestRegressor(random_state=42)

# Define hyperparameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Grid search with cross-validation
grid_search = GridSearchCV(rf, param_grid, cv=5, scoring='r2', n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)

# Best parameters and score
print("Best Parameters:", grid_search.best_params_)
print("Best R² Score:", grid_search.best_score_)

# Train best model on full training data
best_rf = grid_search.best_estimator_
best_rf.fit(X_train, y_train)

y_pred = best_rf.predict(X_test)

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R² Score:", r2_score(y_test, y_pred))
```
